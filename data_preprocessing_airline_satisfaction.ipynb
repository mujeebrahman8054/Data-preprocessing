```python
data['Arrival Delay in Minutes']=data['Arrival Delay in Minutes'].fillna(data['Arrival Delay in Minutes'].median())
data.isna().sum()
```

```python
data.satisfaction.nunique()
```

```python
from sklearn.preprocessing import LabelEncoder
le_satisfaction=LabelEncoder()
le_gender=LabelEncoder()
le_class=LabelEncoder()
```

```python
data['satisfaction_n']=le_satisfaction.fit_transform(data['satisfaction'])
data['gender_n']=le_gender.fit_transform(data['Gender'])
data['class_n']=le_class.fit_transform(data['Class'])
data.head()
```

```python
data_n=data.drop(['satisfaction','Gender','Class'],axis='columns')
```

```python
data_n.head()
```

```python
data_en=data_n.drop(['Customer Type','Type of Travel'],axis='columns')
```

```python
data_en.dtypes
```

```python
y=data_en['satisfaction_n']
X=data_en.drop('satisfaction_n',axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.2)
```

```python
from sklearn import linear_model
lr=linear_model.LogisticRegression()
model=lr.fit(X_train,y_train)
predictions=model.predict(X_test)
predictions
```

```python
y_test
```

```python
model.score(X_test,y_test)
```
